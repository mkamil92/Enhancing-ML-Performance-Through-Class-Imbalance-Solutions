<h1 style="color: indigo;">Function : Confusion matrix</h1>

```python
def plot_cm(cm):
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix')
    plt.show()
```

<h1 style="color: indigo;">Function: Classification Report & Confusion Matric</h1>

```python
def train_test_visualize(model, X, y):
    y_pred = cross_val_predict(model, X, y, cv=5)
    cm = confusion_matrix(y, y_pred)
    plot_cm(cm)
    # Generate classification report
    report = classification_report(y, y_pred, zero_division=1)
    print("Classification Report:\n", report)
```

<h1 style="color: Red;">Baseline Execution</h1>

<h2 style="color: blue;">1- KNN</h2>

```python
def plot_knn(X, y):
  # Define a list of k values to try
  k_values = range(1, 100, 2)

  # Train KNN classifiers with different k values
  scores = list()
  for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    y_pred = cross_val_predict(knn, X, y, cv=5)
    score = accuracy_score(y, y_pred)
    scores.append(score)

  plt.plot(k_values, scores)
  plt.title("Best Value For k")
  plt.xlabel('K')
  plt.ylabel("Accuracy")
  plt.tight_layout()
  plt.show()

  return k_values[scores.index(max(scores))]
k = plot_knn(X, y)
```

```python
def run_knn(X, y):
  knn = KNeighborsClassifier(n_neighbors=k)
  train_test_visualize(knn, X, y)

run_knn(X, y)
```

<h2 style="color: blue;">2- Logistic Regression</h2>

```python
def run_lr(X, y):
  lr = LogisticRegression(max_iter=500)
  train_test_visualize(lr, X, y)

run_lr(X, y)
```

<h2 style="color: blue;">3- Decision Tree Classifier</h2>

```python
from sklearn.tree import DecisionTreeClassifier

def run_tree(X, y):
  tree_model = DecisionTreeClassifier()
  train_test_visualize(tree_model, X, y)

run_tree(X, y)
```

<h2 style="color: blue;">4- SVM</h2>

```python
from sklearn.svm import SVC

def run_svm(X, y):
  # Initialize SVM classifier
  svm_model = SVC(kernel='linear')

  train_test_visualize(svm_model, X, y)

run_svm(X, y)
```

<h2 style="color: blue;">5- Naive Bayes</h2>

```python
from sklearn.naive_bayes import GaussianNB

def run_nb(X, y):
    nb_classifier = GaussianNB()
    train_test_visualize(nb_classifier, X, y)

run_nb(X, y)
```

<h2 style="color:blue">As i have implemented all the techniques and algorithims and before and after imbalance technique with different result so i can conclude results that:<h2>
    <h3 style="color:green">
    -> After applying SMOTE to the given dataset a notable improvements were observed in the performance of all classification algorithms. <br><br>
        -> Before  SMOTE algorithms such as KNN, logistic regression, decision tree, SVM, and Naive Bayes exhibited imbalanced precision and recall for the minority class indicating a struggle in correctly predicting churn instances. But afyer SMOTE these algorithms showed significant enhancements. For exmaple KNN and logistic regression achieved balanced precision, recall, and F1-scores for both classes with accuracy increasing from 80% to 77%. Similarly decision tree and SVM algorithms demonstrated improved precision, recall, and F1-scores for both classes leading to  increase in accuracy from 72% to 77% and 79% to 75% respectively.<br><br>
        -> Naive Bayes also showed improvement with the F1-score for class 1 increasing from 0.61 to 0.76 and the accuracy increasing from 74% to 75%. <br><br>
So we can say that the application of SMOTE effectively addressed the class imbalance in the dataset by enhancing the predictive capabilities of these algorithms for customer churn prediction in the IT industry.
