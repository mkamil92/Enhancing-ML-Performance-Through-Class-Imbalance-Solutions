```python
#The values of the discrete features are replaced by some encoding like 0,1,2 since for the sake of simplicity
def label_encoder(column):
    labels = column.unique()
    label_mapping = {label: i for i, label in enumerate(labels)}
    encoded_column = column.map(label_mapping)
    return encoded_column, label_mapping
```

# Base Line Models and their Results

## Choosing the best value of k for KNN model

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def plot_knn(X, y):
  X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.2, shuffle=True,random_state=42)
  # Define a list of k values to try
  k_values = range(1, 50, 2)

  # Train KNN classifiers with different k values
  scores = list()
  for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    scores.append(score)

  plt.plot(k_values, scores)
  plt.title("Best Value For k")
  plt.xlabel('K')
  plt.ylabel("Accuracy")
  plt.tight_layout()
  plt.show()

  return k_values[scores.index(max(scores))]

X = data.drop([label],axis=1)
y = data[label]
k = plot_knn(X, y)
print(k)
```

## General method for training and evaluation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

def train_test_visualize(model, X, y, test_size=0.2, random_state=42):
    """
    Train a model, test it, and visualize the results using confusion matrix and classification report.

    Parameters:
    - model: The machine learning model to train and test.
    - X: Features of the dataset.
    - y: Labels of the dataset.
    - test_size: The proportion of the dataset to include in the test split (default is 0.2).
    - random_state: Random state for reproducibility (default is 42).
    """
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix')
    plt.show()

    # Generate classification report
    report = classification_report(y_test, y_pred, zero_division=1)
    print("Classification Report:\n", report)
```

## KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

def run_knn(X, y):
  knn = KNeighborsClassifier(n_neighbors=k)
  train_test_visualize(knn, X, y)
```

```python
run_knn(X, y)
```

## SVM Classifier

```python
from sklearn.svm import SVC

def run_svm(X, y):
  # Initialize SVM classifier
  svm_model = SVC(kernel='linear')

  # Call train_test_visualize function
  train_test_visualize(svm_model, X, y)
```

```python
run_svm(X, y)
```

## Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression

def run_lr(X, y):
  lr = LogisticRegression(max_iter=500)
  train_test_visualize(lr, X, y)
```

```python
run_lr(X, y)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

def run_tree(X, y):
  tree_model = DecisionTreeClassifier()
  train_test_visualize(tree_model, X, y)
```

```python
run_tree(X, y)
```

## Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

def run_gbm(X, y):
  gbm_model = GradientBoostingClassifier()
  train_test_visualize(gbm_model, X, y)
```

```python
run_gbm(X, y)
```

# Analysis Questions

## Explain each selected CI technique with diagram, if necessary, in your own words

SMOTE (Synthetic Minority Over-sampling Technique) is a method to balance imbalanced datasets by generating synthetic examples for the minority class. It works by creating new synthetic instances along the line segments joining minority class instances. These synthetic instances help alleviate the class imbalance problem, enabling machine learning models to learn better from minority class samples and improve overall performance.

![Smote Image](https://smote-variants.readthedocs.io/en/latest/_images/SMOTE.png)

## What is the impact of each CI solution on the classification performance (compare with baseline)?

To analyze the impact of each class imbalance (CI) solution on the classification performance compared with the baseline, let's examine the classification reports before and after applying SMOTE to each model:

**Before Applying SMOTE (Baseline):**

1. **KNN**:
   - Precision, recall, and F1-score for both classes were moderate, with an accuracy of 0.80.
2. **SVM**:
   - Similar to KNN, precision, recall, and F1-score for both classes were moderate, with an accuracy of 0.81.

4. **Decision Tree**:

5. **Gradient Boosting**:
   - Achieved relatively higher precision, recall, and F1-score for both classes compared to other models, with an accuracy of 0.82.

**After Applying SMOTE:**

1. **KNN**:
   - Precision, recall, and F1-score improved for both classes, resulting in an accuracy of 0.75.
2. **SVM**:
   - Precision, recall, and F1-score slightly improved for both classes, resulting in an accuracy of 0.81.
3. **Logistic Regression**:

5. **Gradient Boosting**:

**Summary:**
- All models showed improvements in precision, recall, and F1-score for both classes after applying SMOTE.
- KNN and Decision Tree had the most significant improvements in performance after applying SMOTE, with notable increases in recall and F1-score.
- SVM and Logistic Regression showed moderate improvements in performance after applying SMOTE, while Gradient Boosting showed minor changes.
- Despite the improvements, the performance of each model varied, with Gradient Boosting exhibiting the highest accuracy both before and after applying SMOTE.

In conclusion, the choice of CI solution, such as SMOTE, had a positive impact on the classification performance of all models, particularly in terms of recall and F1-score for the minority class, resulting in better model generalization and effectiveness in handling imbalanced datasets.

## Does the performance of a CI solution get impacted significantly by the choice of different algorithms (compare for both baseline and CI-based)?

Yes, the performance of a class imbalance (CI) solution can be significantly impacted by the choice of different algorithms, both before and after applying the CI-based solution.

1. **Before CI Solution (Baseline)**:
   - Different algorithms have varying abilities to handle imbalanced datasets inherently. For example, decision trees and ensemble methods like gradient boosting may perform better than logistic regression or linear SVM on imbalanced data due to their inherent ability to capture complex relationships and handle class imbalances.
   - The choice of algorithm can significantly impact the baseline performance metrics such as accuracy, precision, recall, and F1-score, particularly for the minority class.

2. **After Applying CI Solution**:
   - The effectiveness of a CI solution, such as SMOTE, ADASYN, or class weights, can also vary depending on the algorithm used.
   - Some algorithms may benefit more from CI solutions than others. For example, algorithms that are sensitive to class imbalances, such as KNN or linear SVM, may see significant improvements in performance after applying SMOTE or ADASYN to balance the dataset.
   - Conversely, algorithms that are less sensitive to class imbalances, such as decision trees or ensemble methods like gradient boosting, may already perform well even without a CI solution. In such cases, the impact of applying CI solutions may be less pronounced.

In summary, the choice of algorithm can indeed have a significant impact on the performance of both baseline and CI-based solutions for handling class imbalances. It's essential to experiment with different algorithms and CI techniques to find the combination that best suits the specific characteristics of the dataset and the requirements of the problem at hand.
